```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set style visualisasi
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# --- ASUMSI: df adalah DataFrame kamu ---
# Contoh jika ingin simulasi data (bisa dihapus jika df sudah ada):
# df = pd.read_csv("data_kamu.csv")


# ==========================================
# 1. ANALISIS KOLOM NUMERIK (Skor)
# ==========================================
def analyze_numeric(df, col):
    print(f"=== ANALISIS UNIVARIAT: {col} ===")
    # Statistik Deskriptif
    print(df[col].describe())
    print(f"Missing Values: {df[col].isnull().sum()}\n")

    # Visualisasi (Distribusi & Boxplot)
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Histogram/KDE
    sns.histplot(df[col], kde=True, ax=axes[0], color="skyblue")
    axes[0].set_title(f"Distribusi Frekuensi {col}")

    # Boxplot untuk deteksi outlier
    sns.boxplot(x=df[col], ax=axes[1], color="lightcoral")
    axes[1].set_title(f"Boxplot {col}")

    plt.tight_layout()
    plt.show()


# Run untuk Skor
analyze_numeric(df, "Skor")


# ==========================================
# 2. ANALISIS KOLOM KATEGORIKAL
# (Negara, Tipe Kamar, Tipe Traveler, Evaluasi)
# ==========================================
categorical_cols = ["Negara", "Tipe Kamar", "Tipe Traveler", "Evaluasi"]


def analyze_categorical(df, col):
    print(f"=== ANALISIS UNIVARIAT: {col} ===")
    # Frekuensi dan Persentase
    counts = df[col].value_counts(dropna=False)
    percentages = df[col].value_counts(dropna=False, normalize=True) * 100

    summary = pd.DataFrame({"Jumlah": counts, "Persentase (%)": percentages})
    print(summary)
    print("\n" + "-" * 40 + "\n")

    # Visualisasi (Bar Plot)
    # Mengambil top 10 saja jika datanya terlalu banyak (seperti Negara)
    top_n = df[col].value_counts().head(10)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_n.values, y=top_n.index, palette="viridis")
    plt.title(f"Top Frekuensi Kolom: {col}")
    plt.xlabel("Jumlah")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()


# Run untuk semua kolom kategorikal
for col in categorical_cols:
    if col in df.columns:
        analyze_categorical(df, col)


# ==========================================
# 3. ANALISIS KOLOM TANGGAL (Tanggal Ulasan)
# ==========================================
print("=== ANALISIS UNIVARIAT: Tanggal Ulasan ===")
# Ubah ke datetime (sesuaikan format jika diperlukan)
df["Tanggal Ulasan_parsed"] = pd.to_datetime(
    df["Tanggal Ulasan"], errors="coerce"
)

print(f"Rentang Tanggal: {df['Tanggal Ulasan_parsed'].min()} s/d {df['Tanggal Ulasan_parsed'].max()}")
print(f"Missing/Invalid Dates: {df['Tanggal Ulasan_parsed'].isnull().sum()}\n")

# Tren Berdasarkan Bulan/Tahun
df["Tahun_Bulan"] = df["Tanggal Ulasan_parsed"].dt.to_period("M")
trend_data = df["Tahun_Bulan"].value_counts().sort_index()

if not trend_data.empty:
    plt.figure(figsize=(12, 5))
    trend_data.plot(kind="line", marker="o", color="green")
    plt.title("Tren Jumlah Ulasan Per Bulan")
    plt.xlabel("Waktu (Tahun-Bulan)")
    plt.ylabel("Jumlah Ulasan")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# ==========================================
# 4. ANALISIS KOLOM TEKS / CARDINALITY TINGGI
# (Nama, Detail Menginap, Judul, Komentar)
# ==========================================
text_cols = ["Nama", "Detail Menginap", "Judul", "Komentar"]

print("=== ANALISIS UNIVARIAT: Kolom Teks & Kardinalitas Tinggi ===")
for col in text_cols:
    if col in df.columns:
        unique_counts = df[col].nunique()
        missing = df[col].isnull().sum()
        print(f"Kolom [{col}]:")
        print(f"  - Jumlah Unik: {unique_counts} (dari 302 data)")
        print(f"  - Missing Values: {missing}")

        # Khusus untuk Judul dan Komentar, kita cek panjang karakternya
        if col in ["Judul", "Komentar"]:
            # Isi missing value dengan string kosong untuk hitung panjang teks
            text_lengths = df[col].fillna("").apply(len)
            print(f"  - Rata-rata Panjang Karakter: {text_lengths.mean():.2f}")
            print(f"  - Panjang Karakter Maksimal: {text_lengths.max()}")

            # Plot panjang karakter
            plt.figure(figsize=(8, 3))
            sns.histplot(text_lengths, bins=20, color="orange", kde=True)
            plt.title(f"Distribusi Panjang Karakter pada {col}")
            plt.xlabel("Jumlah Karakter")
            plt.show()
        print("-" * 30)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set style visualisasi
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# --- ASUMSI: df adalah DataFrame kamu ---
# df = pd.read_csv("data_kamu.csv")

# ==========================================
# 0. PERSIAPAN DATA (FEATURE ENGINEERING)
# ==========================================
# Membuat kolom numerik baru dari panjang komentar
df["Panjang_Komentar"] = df["Komentar"].fillna("").apply(len)

# Konversi tanggal
df["Tanggal_Parsed"] = pd.to_datetime(df["Tanggal Ulasan"], errors="coerce")
df["Tahun_Bulan"] = df["Tanggal_Parsed"].dt.to_period("M")


# ==========================================
# 1. ANALISIS BIVARIAT (Hubungan 2 Variabel)
# ==========================================
print("=== ANALISIS BIVARIAT ===")

# a. Kategorikal vs Numerik (Tipe Traveler vs Skor)
plt.figure(figsize=(10, 6))
sns.boxplot(x="Skor", y="Tipe Traveler", data=df, palette="Set2")
plt.title("Distribusi Skor Berdasarkan Tipe Traveler")
plt.xlabel("Skor")
plt.ylabel("Tipe Traveler")
plt.tight_layout()
plt.show()

# b. Kategorikal vs Numerik (Evaluasi vs Skor)
plt.figure(figsize=(10, 6))
sns.violinplot(x="Evaluasi", y="Skor", data=df, palette="pastel")
plt.title("Distribusi Skor Berdasarkan Label Evaluasi")
plt.xlabel("Evaluasi")
plt.ylabel("Skor")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# c. Numerik vs Numerik (Panjang Komentar vs Skor)
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Panjang_Komentar", y="Skor", data=df, alpha=0.6, color="royalblue")
plt.title("Hubungan Panjang Komentar dengan Skor")
plt.xlabel("Panjang Karakter Komentar")
plt.ylabel("Skor")
plt.tight_layout()
plt.show()

# d. Waktu vs Numerik (Tren Skor Rata-rata per Bulan)
tren_skor = df.groupby("Tahun_Bulan")["Skor"].mean()
if not tren_skor.empty:
    plt.figure(figsize=(12, 5))
    tren_skor.plot(kind="line", marker="o", color="darkorange", linewidth=2)
    plt.title("Tren Rata-rata Skor per Bulan")
    plt.xlabel("Waktu (Tahun-Bulan)")
    plt.ylabel("Rata-rata Skor")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# ==========================================
# 2. ANALISIS MULTIVARIAT (Hubungan >= 3 Variabel)
# ==========================================
print("\n=== ANALISIS MULTIVARIAT ===")

# a. Numerik vs Kategorikal vs Kategorikal (Skor vs Tipe Traveler & Evaluasi)
# Melihat skor berdasarkan Tipe Traveler, dipisah lagi warnanya berdasarkan Evaluasi
plt.figure(figsize=(14, 7))
sns.barplot(x="Tipe Traveler", y="Skor", hue="Evaluasi", data=df, errorbar=None, palette="muted")
plt.title("Rata-rata Skor Berdasarkan Tipe Traveler dan Evaluasi")
plt.xlabel("Tipe Traveler")
plt.ylabel("Rata-rata Skor")
plt.legend(title="Evaluasi", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# b. Numerik vs Numerik vs Kategorikal (Panjang Komentar vs Skor dipisah Tipe Traveler)
plt.figure(figsize=(12, 7))
sns.scatterplot(x="Panjang_Komentar", y="Skor", hue="Tipe Traveler", data=df, alpha=0.7, palette="tab10")
plt.title("Hubungan Panjang Komentar dan Skor Berdasarkan Tipe Traveler")
plt.xlabel("Panjang Karakter Komentar")
plt.ylabel("Skor")
plt.legend(title="Tipe Traveler", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# c. Korelasi antar variabel numerik (Skor & Panjang_Komentar)
# Jika ada lebih dari 2 numerik, Heatmap korelasi sangat berguna
numeric_cols = ["Skor", "Panjang_Komentar"]
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
plt.title("Heatmap Korelasi Variabel Numerik")
plt.tight_layout()
plt.show()

# d. Pairplot untuk melihat distribusi dan hubungan antar variabel numerik sekaligus
# Tambahkan 'hue' untuk melihat dimensi kategorikal
sns.pairplot(df[["Skor", "Panjang_Komentar", "Tipe Traveler"]], hue="Tipe Traveler", palette="husl", corner=True)
plt.suptitle("Pairplot Variabel Numerik Berdasarkan Tipe Traveler", y=1.02)
plt.show()
```
